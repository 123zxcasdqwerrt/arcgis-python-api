## Conda Environments
Now you have conda and the ArcGIS API for Python installed. Let's talk about the basics of conda so you can manage installations and set up IDEs for projects and scripts.

The easiest way to understand conda is by working through the [30-minute Getting Started](https://conda.io/docs/user-guide/getting-started.html) tutorial.

Conda manages packages and dependencies through [environments](https://conda.io/docs/user-guide/concepts.html#conda-environments), which are physical directories on your disk system that isolate the specific Python interpreter and packages you install in them. You can easily create and switch between environments without affecting other environments. For detailed explanation and instructions, see the conda documentation on [Managing environments](https://conda.io/docs/user-guide/tasks/manage-environments.html). 

## The Default Environments:
You can list the conda environments on your machine in a terminal application:

`conda env list`

**NOTE:** Add the directory with the conda executables to your PATH to run the conda command from anywhere.

## Activating Environments
Use the activate command to switch between environments at a terminal application. See [Activating an environment](https://conda.io/docs/user-guide/tasks/manage-environments.html#activating-an-environment) for more information:
 * **Windows**:
 
    `activate <environment_name>`
<p></p>    
 * **macOS and Linux**: 
 
    `source activate <environment_name>`

## Adding Packages to Environments
The conda package management system can install thousands of packages accessed from various repositories called Channels. Channels may be a default set of cloud locations, such as Anaconda Cloud, or private repositories created by individuals or organizations. Conda searches a default set of channels when listing or installing packages. See [Managing Channels](https://conda.io/docs/user-guide/tasks/manage-channels.html) for more information.

Esri has created a conda repository at a channel named esri to store the arcgis package, which is how you installed the ArcGIS API for Python.

You can add a package to the currently active environment at the terminal with the following command:

`conda install <package_name>`

See [Managing Packages](https://conda.io/docs/user-guide/tasks/manage-pkgs.html) for detailed instructions on using the conda command-line interface. See the [The Default Environments](#The-Default-Environments) section for detailed instructions on using ArcGIS Pro or Anaconda tools.


## Setting up IDEs
To begin projects in IDEs, you often have to specify the path to the Python interpreter. When using conda environments, each environment you uses a different Python interpreter. Type the following to get a list of all environments on your machine and the directory location of the specific Python interpreter (python.exe) used for that interpreter:

`conda info --envs`

![conda_info_envs Image Goes Here](http://esri.github.io/arcgis-python-api/notebooks/nbimages/conda_info_envs.png)

Enter the directory name and add python.exe to set the correct path to the Python interpreter for a specific environment.For example, the correct interpreter for the arcgispro-py3 environment in an IDE based on the screenshot above would be:

`C:\Program Files\ArcGIS\Pro\bin\Python\envs\arcgispro-py3\python.exe`

See [Finding your Anaconda Python interpreter path](https://docs.anaconda.com/anaconda/user-guide/tasks/integration/python-path) for more detailed instructions on different systems.

## Step by Step on macOS: A GIS Environment

## Create an environment:
```bash
conda create --name gis-enviro
```

> You can name your environment anything you want.

## Install the ArcGIS API for PythonÂ¶

```bash
conda install --name gis-enviro -c esri arcgis
```
**Note**: if you do not enter an environment name, the package will be installed in the root environment.

## Activate the environment

```bash
source activate gis-enviro
```
