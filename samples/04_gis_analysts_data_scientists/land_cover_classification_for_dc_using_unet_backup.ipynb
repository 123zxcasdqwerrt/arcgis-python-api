```python
from arcgis.learn import UnetClassifier, prepare_data
```

```python
data_path = r'C:\Users\yong9856\Downloads\land_cover_projected_new_clip'
```

```python
# codes = {0:'NoData', 1: 'Water', 2: 'Wetlands', 3: 'Tree1', 4: 'Shurbland', 5: 'Lowvegetation', 6: 'Barren', 7: 'Structures', 8:'Impervious',
#        9: 'Roads', 10: 'Tree2', 11: 'Tree3', 11: 'Tree4'}
```

```python
data = prepare_data(data_path, batch_size=4)
```

```python
data.show_batch()
```

```python
model = UnetClassifier(data)
```

```python
model.lr_find()
```

```python
model.fit(50, lr=slice(1e-4, 5e-4))
```

```python
model.show_results()
```

```python
model.save('50-epoch')
```
