```python
from arcgis.learn import UnetClassifier, prepare_data
```

```python
data_path = r'C:\Users\yong9856\Downloads\land_cover_kent'
```

```python
data = prepare_data(data_path, batch_size=4)
```

```python
data
```

```python
data.show_batch()
```

```python
model = UnetClassifier(data)
```

```python
model.lr_find()
```

```python
model.fit(10, lr=slice(1e-4, 1e-3))
```

```python
model.save('stage-1-10')
```

```python
model.load('stage-1-10');
```
