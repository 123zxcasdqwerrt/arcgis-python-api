## Introduction

We can use `arcgis.learn` to train deep learning models that can can be deployed on mobile devices using ArcGIS field apps. This enables AI on the edge and simplify jobs of field workers. This is done using Tensorflow which enables us to save trained models to tflite format. A few applications of this workflow are 
- Assisted intelligent survey,
- Disconnected Descision support,
- Automatic feature extraction from street view,
- Automated Rapid data collection on the move,


 In this sample we are going to use street view data from ESRI campus, we have labeled a few fire water hydrants. 

image.png

# Prerequisites

- This workflow is a build up on top of the basic workflow followed to train deep learning models with `arcgis.learn` module. It is recommended to first have a reading of the basic workflow from [here](#https://developers.arcgis.com/python/guide/geospatial-deep-learning/) if not done previously. 
- To train mobile ready light models **tensorflow** needs to be installed in addition to fastai and pytorch, please refer to the section "Install deep learning dependencies of arcgis.learn module" [on this page](https://developers.arcgis.com/python/guide/install-and-set-up/) for detailed documentation on installation of these dependencies.


## Supported Models

Currently the following models can be used to deploy on mobile devices.  
- SingleShotDetector
- UnetClassifier
- FeatureClassifier

## Set Environment Variables

```python
%env ARCGIS_ENABLE_TF_BACKEND=1
```

## Imports

```python
from arcgis.learn import prepare_data, SingleShotDetector
```

```python
data = prepare_data(
    'Esri Campus Street view',
    dataset_type='PASCAL_VOC_rectangles',
    batch_size=16
)
```

## Visualize Training data

```python
data.show_batch()
```

## Train Model

### Model instantiation options


 - `backbone`: Make sure you pick a mobile optimised backbone, currently we only support 'MobileNetV2' backbone.
 - `backend`: **Important !**, Make sure you select the backend as 'tensorflow' to enable saving the model to tflite format.

```python
model = SingleShotDetector(data, backbone='MobileNetV2', backend='tensorflow')
```

We would now find an appropirate learning rate using the `{model}.lr_find()` method.

```python
model.lr_find()
```

We would now train the model by calling the `{model}.fit()` method.

```python
model.fit(10, .001)
```

One our model has been fitted for the specified epochs, we can visualize a few sample predictions from the model using the `{model}.show_results()` method.

## Save Model to an intermediate format 

We can use the `{model}.save()` mwthod to save the model to 'keras_saved_model' format, this format can be used to save and load intermediate models in the training workflow. 

Note: **Important!** This format is not directly deployable on mobile devices and needs to exported to mobile ready format as explained in [this step](#Exporting-model-to-mobile-ready-format). 

```python
model.save('10e')
```

Visualizing a few results 

```python
model.show_results(thresh=0.5)
```

If you want to load a model save to an intermediate format explained [here](#Save-Model-to-an-intermediate-format) you can use the `{model}.load()` method.

```python
# model.load('10e')
```

## Exporting model to mobile ready format

We will now save the model we just trained to a format supported by Esri Field Apps. Right now the only supported format is 'tflite', and it is required to set the parameter'framework'='tflite' while calling the `{model}.save()` method.

Note: **Important !** This is an export only format, and once exported the output cannot be loaded back to the training workflow and fine tuned again. To save checkpoints or reusable models that can be loaded back, you need to save the model to a intermediate format as explained [here](#Save-Model-to-an-intermediate-format)

```python
model.save('10e_tflite', framework="tflite")
```
