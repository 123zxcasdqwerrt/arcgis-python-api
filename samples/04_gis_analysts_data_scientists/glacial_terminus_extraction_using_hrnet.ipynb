## Introduction

With the change in global climate, glaciers all over the world are experiencing an increasing mass loss, resulting in changing calving fronts. This calving front delineation is important to monitor the rate of glacial mass loss. Currently, most of the calving front delineation is done manually resulting in huge time consumption and under-utilization of satellite imagery.

Extracting calving fronts from satellite images of marine-terminating glaciers is a two-step process. First, segmenting the Front using different segmentation techniques, and second, post-processing mechanisms to extract the terminus line. This notebook presents the use of HRNet model from `arcgis.learn` module to accomplish the first task of segmenting calving fronts. We have used the data provided in [CALFIN](https://github.com/daniel-cheng/CALFIN) repository. The training data includes 1600+ Greenlandic glaciers and 200+ Antarctic glaciers/ice shelves images from Landsat (optical) and Sentinel-1 (SAR) satellites.

## Necessary imports

```python
import os
import glob
import zipfile
from pathlib import Path

from arcgis.gis import GIS
from arcgis.learn import MMSegmentation, prepare_data
```

## Connect to your GIS

```python
# Connect to GIS
gis = GIS("home")
```

## Download training data 

```python
training_data = gis.content.get('cc750295180a487aa7af67a67cadff78')
training_data
```

The data size is around 6.5 GBs and it may take some time to get downloaded.

```python
filepath = training_data.download(file_name=training_data.name)
```

```python
with zipfile.ZipFile(filepath, 'r') as zip_ref:
    zip_ref.extractall(Path(filepath).parent)
```

```python
output_path = os.path.join(os.path.splitext(filepath)[0])
```

```python
output_path = glob.glob(output_path)
```

## Train the model

`arcgis.learn` provides HRNet model through the integration of `MMSegmentation` class. To have in-depth information about MMSegmentation's working and usage, have a look at the guide - [Using MMSegmentation with arcgis.learn](https://developers.arcgis.com/python/guide/using-mmsegmentation-with-arcgis-learn/).

### Prepare data

We will specify the path to our training data and a few hyperparameters.

- `path`: path of the folder/list of folders containing training data.
- `batch_size`: Number of images your model will train on each step inside an epoch, it directly depends on the memory of your graphic card.

```python
data = prepare_data(path=output_path, dataset_type='Classified_Tiles', batch_size=4)
```

### Visualize training data

To get a sense of what the training data looks like, `arcgis.learn.show_batch()` method randomly picks a few training chips and visualizes them.
- `rows`: Number of rows to visualize

```python
data.show_batch(5, alpha=0.7)
```

### Load model architecture

```python
model = MMSegmentation(data, 'hrnet')
```

### Find an optimal learning rate

[Learning rate](https://en.wikipedia.org/wiki/Stochastic_gradient_descent) is one of the most important hyperparameters in model training. `ArcGIS API for Python` provides a learning rate finder that automatically chooses the optimal learning rate for you.

```python
lr = model.lr_find()
```

### Fit the model 

We will train the model for a few epochs with the learning rate we have found. For the sake of time, we can start with 30 epochs.

```python
model.fit(30, lr)
```

As we can see train and valid losses are still going down validating that the model is still learning. There is more room for training, so we trained the model for 170 epochs in total to achieve good results.

### Visualize results in validation set

It is a good practice to see the results of the model viz-a-viz ground truth. The code below picks random samples and shows us ground truth and model predictions, side by side. This enables us to preview the results of the model we trained for 170 epochs within the notebook.

```python
model.show_results(5, thresh=0.1, aplha=0.1)
```

### Accuracy assessment

`arcgis.learn` provides `mIOU()` that computes mean IOU on the validation set for each class.

```python
model.mIOU()
```

### Save the model

We will save the model which we trained as a 'Deep Learning Package' ('.dlpk' format). The Deep Learning package is the standard format used to deploy deep learning models on the ArcGIS platform.

We will use the `save()` method to save the trained model. By default, it will be saved to the 'models' sub-folder within our training data folder.

```python
model.save("170e_hrnet_Glaciertips")
```

## Model inference

In this step, we will generate a classified raster using 'Classify Pixels Using Deep Learning' tool available in both `ArcGIS Pro` and `ArcGIS Enterprise`.

- `Input Raster`: The raster layer you want to classify.
- `Model Definition`: It will be located inside the saved model in the 'models' folder in '.emd' format.
- `Padding`: The 'Input Raster' is tiled and the deep learning model classifies each individual tile separately before producing the final 'Output Classified Raster'. This may lead to unwanted artifacts along the edges of each tile as the model has little context to predict accurately. Padding allows us to supply some extra information along the tile edges, this helps the model to predict better.
- `Cell Size`: Should be close to the size used to train the model.
- `Processor Type`: This allows you to control whether the system's 'GPU' or 'CPU' will be used to classify pixels, by 'default GPU' will be used if available.
